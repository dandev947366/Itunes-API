Itunes Search API

```python
import requests
import json
import pandas as pd
```

```python
baseUrl = "https://itunes.apple.com/search"
```

```python
r = requests.get(baseUrl, params = {
    "term": "the beatles",
    "country": "us",
    "limit": 200
})

r.status_code
```

```python
info = r.json()
```

```python
print(json.dumps(info, indent=2))
```

```python
# detail of first result
print(json.dumps(info["results"][0], indent=2))
```

```python
# find name of artist
for result in info["results"]:
    print(result["artistName"])
```

```python
# release date
info["results"][0]["releaseDate"]
```

```python
for result in info["results"]:
    print(result["releaseDate"])
```

Structuring and Export Data

```python
songs_df = pd.DataFrame(info["results"])

print(songs_df)
```

```python
# save to csv file
songs_df.to_csv("song_info.csv")
```
